**[From Leetcode: 18. 4Sum](https://leetcode.com/problems/4sum)**

<div class="xFUwe" data-track-load="description_content"><p>Given an array <code>nums</code> of <code>n</code> integers, return <em>an array of all the <strong>unique</strong> quadruplets</em> <code>[nums[a], nums[b], nums[c], nums[d]]</code> such that:</p>

<ul>
	<li><code>0 &lt;= a, b, c, d&nbsp;&lt; n</code></li>
	<li><code>a</code>, <code>b</code>, <code>c</code>, and <code>d</code> are <strong>distinct</strong>.</li>
	<li><code>nums[a] + nums[b] + nums[c] + nums[d] == target</code></li>
</ul>

<p>You may return the answer in <strong>any order</strong>.</p>

<p>&nbsp;</p>
<p><strong class="example">Example 1:</strong></p>

<pre><strong>Input:</strong> nums = [1,0,-1,0,-2,2], target = 0
<strong>Output:</strong> [[-2,-1,1,2],[-2,0,0,2],[-1,0,0,1]]
</pre>

<p><strong class="example">Example 2:</strong></p>

<pre><strong>Input:</strong> nums = [2,2,2,2,2], target = 8
<strong>Output:</strong> [[2,2,2,2]]
</pre>

<p>&nbsp;</p>
<p><strong>Constraints:</strong></p>

<ul>
	<li><code>1 &lt;= nums.length &lt;= 200</code></li>
	<li><code>-10<sup>9</sup> &lt;= nums[i] &lt;= 10<sup>9</sup></code></li>
	<li><code>-10<sup>9</sup> &lt;= target &lt;= 10<sup>9</sup></code></li>
</ul>
</div>

```
def fourSum(nums, target):
    arr = {}
    nums = sorted(nums)

    for ix in range(len(nums)):
        for i in range(ix + 1, len(nums)):
            j = i + 1
            k = len(nums) - 1
            while j < k:
                sum_ = nums[ix] + nums[i] + nums[j] + nums[k]
                if sum_ == target:
                    arr['|'.join([str(nums[ix]), str(nums[i]), str(nums[j]), str(nums[k])])] = ''
                if sum_ > target:
                    k -= 1
                else:
                    j += 1

    return [[int(n) for n in key.split('|')] for key in arr.keys()]


fourSum(nums = [4,0,5,-5,3,3,0,-4,-5], target = -2)
```
