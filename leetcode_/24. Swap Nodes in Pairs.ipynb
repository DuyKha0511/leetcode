**[From Leetcode: 24. Swap Nodes in Pairs](https://leetcode.com/problems/swap-nodes-in-pairs)**

<div class="xFUwe" data-track-load="description_content"><p>Given a&nbsp;linked list, swap every two adjacent nodes and return its head. You must solve the problem without&nbsp;modifying the values in the list's nodes (i.e., only nodes themselves may be changed.)</p>

<p>&nbsp;</p>

<p><strong>Constraints:</strong></p>

<ul>
	<li>The number of nodes in the&nbsp;list&nbsp;is in the range <code>[0, 100]</code>.</li>
	<li><code>0 &lt;= Node.val &lt;= 100</code></li>
</ul>
</div>

```
# Definition for singly-linked list.
# class ListNode(object):
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution(object):
    def swapPairs(self, head):
        """
        :type head: ListNode
        :rtype: ListNode
        """
        temp = head
        flag = 0
        node_temp = None

        while temp is not None:
            node1 = temp
            node2 = temp.next
            if node2 == None:
                break

            node1.next = node2.next
            node2.next = node1

            if node_temp is not None:
                node_temp.next = node2

            node_temp = node1

            temp = node2
            if flag == 0:
                head = temp
                flag = 1

            temp = temp.next
            temp = temp.next

        return head
```
