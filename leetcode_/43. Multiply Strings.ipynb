**[From Leetcode: 43. Multiply Strings](https://leetcode.com/problems/multiply-strings)**

<div class="xFUwe" data-track-load="description_content"><p>Given two non-negative integers <code>num1</code> and <code>num2</code> represented as strings, return the product of <code>num1</code> and <code>num2</code>, also represented as a string.</p>

<p><strong>Note:</strong>&nbsp;You must not use any built-in BigInteger library or convert the inputs to integer directly.</p>

<p>&nbsp;</p>
<p><strong class="example">Example 1:</strong></p>
<pre><strong>Input:</strong> num1 = "2", num2 = "3"
<strong>Output:</strong> "6"
</pre><p><strong class="example">Example 2:</strong></p>
<pre><strong>Input:</strong> num1 = "123", num2 = "456"
<strong>Output:</strong> "56088"
</pre>
<p>&nbsp;</p>
<p><strong>Constraints:</strong></p>

<ul>
	<li><code>1 &lt;= num1.length, num2.length &lt;= 200</code></li>
	<li><code>num1</code> and <code>num2</code> consist of digits only.</li>
	<li>Both <code>num1</code> and <code>num2</code>&nbsp;do not contain any leading zero, except the number <code>0</code> itself.</li>
</ul>
</div>

```
class Solution:
    def multiply(self, num1: str, num2: str) -> str:
        """"""
        if len(num2) > len(num1):
            num2, num1 = num1, num2
        
        result = 0
        x10 = 1
        for i in range(len(num2) - 1, -1, -1):
            print(self.multiply_digit(num1, num2[i], x10))
            result += self.multiply_digit(num1, num2[i], x10)
            x10 *= 10
        
        return result
        
        
    def multiply_digit(self, number, digit, x10_):
        if digit == '0':
            return 0
        
        if digit == '1':
            return int(number) * x10_
        
        digit = int(digit)
        result = 0
        x10 = 1
        for i in range(len(number) - 1, -1, -1):
            result += digit * int(number[i]) * x10
            x10 *= 10
        
        return result * x10_
    
s = Solution()
s.multiply('42', '10')
```
