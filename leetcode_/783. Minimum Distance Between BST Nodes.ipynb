<div class="xFUwe" data-track-load="description_content"><p>Given the <code>root</code> of a Binary Search Tree (BST), return <em>the minimum difference between the values of any two different nodes in the tree</em>.</p>

</ul>

<p>&nbsp;</p>
<p><strong>Note:</strong> This question is the same as 530: <a href="https://leetcode.com/problems/minimum-absolute-difference-in-bst/" target="_blank">https://leetcode.com/problems/minimum-absolute-difference-in-bst/</a></p>

```
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution(object):
    def minDiffInBST(self, root):
        """
        :type root: TreeNode
        :rtype: int
        """
        nums = self.find(root, [])
        min_ = 2**31
        for i in range(1, len(nums)):
            if nums[i] - nums[i - 1] <= min_:
                min_ = nums[i] - nums[i - 1]

        return min_

    def find(self, root, array):
        if root != None:
            self.find(root.left, array)
            array.append(root.val)
            self.find(root.right, array)

        return array
```
