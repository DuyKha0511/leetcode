**[From Leetcode: 589. N-ary Tree Preorder Traversal](https://leetcode.com/problems/n-ary-tree-preorder-traversal)**

<div class="xFUwe" data-track-load="description_content"><p>Given the <code>root</code> of an n-ary tree, return <em>the preorder traversal of its nodes' values</em>.</p>

<p>Nary-Tree input serialization is represented in their level order traversal. Each group of children is separated by the null value (See examples)</p>

	<li><code>0 &lt;= Node.val &lt;= 10<sup>4</sup></code></li>

<p>&nbsp;</p>
<p><strong>Follow up:</strong> Recursive solution is trivial, could you do it iteratively?</p>

```
"""
# Definition for a Node.
class Node(object):
    def __init__(self, val=None, children=None):
        self.val = val
        self.children = children
"""

class Solution(object):
    def preorder(self, root):
        """
        :type root: Node
        :rtype: List[int]
        """
        return self.pre(root, [])

    def pre(self, root, paths):
        if root is not None:
            paths.append(root.val)
            for child in root.children:
                self.pre(child, paths)


        return paths
```
