**[From Leetcode: 999. Available Captures for Rook](https://leetcode.com/problems/available-captures-for-rook)**

<div class="xFUwe" data-track-load="description_content"><p>On an <code>8 x 8</code> chessboard, there is <strong>exactly one</strong> white rook <code>'R'</code> and some number of white bishops <code>'B'</code>, black pawns <code>'p'</code>, and empty squares <code>'.'</code>.</p>

<p>When the rook moves, it chooses one of four cardinal directions (north, east, south, or west), then moves in that direction until it chooses to stop, reaches the edge of the board, captures a black pawn, or is blocked by a white bishop. A rook is considered <strong>attacking</strong> a pawn if the rook can capture the pawn on the rook's turn. The <strong>number of available captures</strong> for the white rook is the number of pawns that the rook is <strong>attacking</strong>.</p>

<p>Return <em>the <strong>number of available captures</strong> for the white rook</em>.</p>

<p><strong>Constraints:</strong></p>

<ul>
	<li><code>board.length == 8</code></li>
	<li><code>board[i].length == 8</code></li>
	<li><code>board[i][j]</code> is either <code>'R'</code>, <code>'.'</code>, <code>'B'</code>, or <code>'p'</code></li>
	<li>There is exactly one cell with <code>board[i][j] == 'R'</code></li>
</ul>
</div>

```
class Solution(object):
    def numRookCaptures(self, board):
        """
        :type board: List[List[str]]
        :rtype: int
        """
        x_rook = 0
        y_rook = 0
        for i in range(8):
            flag = False
            for j in range(8):
                if board[i][j] == 'R':
                    x_rook, y_rook = i, j
                    flag = True
                    break

            if flag: break
        n_captures = 0
        for i in range(x_rook, -1, -1):
            if board[i][y_rook] == 'p':
                n_captures += 1
                break
            elif board[i][y_rook] == 'B':
                break

        for i in range(x_rook, 8):
            if board[i][y_rook] == 'p':
                n_captures += 1
                break
            elif board[i][y_rook] == 'B':
                break

        for i in range(y_rook, -1, -1):
            if board[x_rook][i] == 'p':
                n_captures += 1
                break
            elif board[x_rook][i] == 'B':
                break

        for i in range(y_rook, 8):
            if board[x_rook][i] == 'p':
                n_captures += 1
                break
            elif board[x_rook][i] == 'B':
                break

        return n_captures

```
