**[From Leetcode: 1360. Number of Days Between Two Dates](https://leetcode.com/problems/number-of-days-between-two-dates)**

<div class="xFUwe" data-track-load="description_content"><p>Write a program to count the number of days between two dates.</p>

<p>The two dates are given as strings, their format is <code>YYYY-MM-DD</code>&nbsp;as shown in the examples.</p>

<p>&nbsp;</p>
<p><strong class="example">Example 1:</strong></p>
<pre><strong>Input:</strong> date1 = "2019-06-29", date2 = "2019-06-30"
<strong>Output:</strong> 1
</pre><p><strong class="example">Example 2:</strong></p>
<pre><strong>Input:</strong> date1 = "2020-01-15", date2 = "2019-12-31"
<strong>Output:</strong> 15
</pre>
<p>&nbsp;</p>
<p><strong>Constraints:</strong></p>

<ul>
	<li>The given dates are valid&nbsp;dates between the years <code>1971</code> and <code>2100</code>.</li>
</ul>
</div>

```
class Solution(object):
    def daysBetweenDates(self, date1, date2):
        n_days1 = self.dayOfYear(date1)
        year1 = int(date1.split('-')[0])
        n_days2 = self.dayOfYear(date2)
        year2 = int(date2.split('-')[0])

        if year1 == year2:
            return abs(n_days1 - n_days2)

        if year1 > year2:
            year1, year2 = year2, year1
            n_days1, n_days2 = n_days2, n_days1

        n_days = n_days2
        for year in range(year1, year2):
            n_days += 366 if self.is_leap_year(year) else 365

        return n_days - n_days1


    def dayOfYear(self, date):
        day_of_months = [
            31,
            28,
            31,
            30,
            31,
            30,
            31,
            31,
            30,
            31,
            30,
            31
        ]
        date = date.split('-')
        year, month, day = int(date[0]), int(date[1]), int(date[2])

        if self.is_leap_year(year):
            day_of_months[1] += 1

        for i in range(1, month):
            day += day_of_months[i-1]

        return day

    def is_leap_year(self, year):
        if year % 400 == 0:
            return True

        if year % 4 == 0 and year % 100 != 0:
            return True

        return False
```
