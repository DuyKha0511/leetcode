**[From Leetcode: 993. Cousins in Binary Tree](https://leetcode.com/problems/cousins-in-binary-tree)**

<div class="xFUwe" data-track-load="description_content"><p>Given the <code>root</code> of a binary tree with unique values and the values of two different nodes of the tree <code>x</code> and <code>y</code>, return <code>true</code> <em>if the nodes corresponding to the values </em><code>x</code><em> and </em><code>y</code><em> in the tree are <strong>cousins</strong>, or </em><code>false</code><em> otherwise.</em></p>

<p>Two nodes of a binary tree are <strong>cousins</strong> if they have the same depth with different parents.</p>

<p>Note that in a binary tree, the root node is at the depth <code>0</code>, and children of each depth <code>k</code> node are at the depth <code>k + 1</code>.</p>

	<li>The number of nodes in the tree is in the range <code>[2, 100]</code>.</li>
	<li><code>1 &lt;= Node.val &lt;= 100</code></li>
	<li>Each node has a <strong>unique</strong> value.</li>
	<li><code>x != y</code></li>
	<li><code>x</code> and <code>y</code> are exist in the tree.</li>

```
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution(object):
    def isCousins(self, root, x, y):
        """
        :type root: TreeNode
        :type x: int
        :type y: int
        :rtype: bool
        """
        sum_, parent = self.get_all_nodes_by_level(root, 0)
        parent[root.val] = -1
        if parent[x] == parent[y]: return False
        if sum_[x] != sum_[y]: return False

        return True

    def get_all_nodes_by_level(self, root, level = 0, sum_ = {}, parent = {}):
        if root is not None:
            sum_[root.val] = level
            if root.left is not None:
                parent[root.left.val] = root.val
                sum_, parent = self.get_all_nodes_by_level(root.left, level + 1, sum_, parent)
            if root.right is not None:
                parent[root.right.val] = root.val
                sum_, parent = self.get_all_nodes_by_level(root.right, level + 1, sum_, parent)

        return sum_, parent
```
