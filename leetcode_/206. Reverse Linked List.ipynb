**[From Leetcode: 206. Reverse Linked List](https://leetcode.com/problems/reverse-linked-list)**

<div class="xFUwe" data-track-load="description_content"><p>Given the <code>head</code> of a singly linked list, reverse the list, and return <em>the reversed list</em>.</p>

<p>&nbsp;</p>

<p><strong>Constraints:</strong></p>

<ul>
	<li>The number of nodes in the list is the range <code>[0, 5000]</code>.</li>
	<li><code>-5000 &lt;= Node.val &lt;= 5000</code></li>
</ul>

<p>&nbsp;</p>
<p><strong>Follow up:</strong> A linked list can be reversed either iteratively or recursively. Could you implement both?</p>
</div>

```
class ListNode(object):
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def reverseList(head):
    if head == None: return None
    temp = head
    reverse_list = None
    while temp:
        new_node = ListNode(temp.val, reverse_list)
        reverse_list = new_node
        temp = temp.next
    return reverse_list
```
