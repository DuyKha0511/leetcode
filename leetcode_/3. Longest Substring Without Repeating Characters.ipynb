**[From Leetcode: 3. Longest Substring Without Repeating Characters](https://leetcode.com/problems/longest-substring-without-repeating-characters)**

<div class="xFUwe" data-track-load="description_content"><p>Given a string <code>s</code>, find the length of the <strong>longest</strong> <span data-keyword="substring-nonempty" class=" cursor-pointer relative text-dark-blue-s text-sm"><div class="popover-wrapper inline-block" data-headlessui-state=""><div><div aria-expanded="false" data-headlessui-state="" id="headlessui-popover-button-:rf:"><div><strong>substring</strong></div></div><div style="position: fixed; z-index: 40; inset: 0px auto auto 0px; transform: translate(354px, 248px);"></div></div></div></span> without repeating characters.</p>

<p>&nbsp;</p>
<p><strong class="example">Example 1:</strong></p>

<pre><strong>Input:</strong> s = "abcabcbb"
<strong>Output:</strong> 3
<strong>Explanation:</strong> The answer is "abc", with the length of 3.
</pre>

<p><strong class="example">Example 2:</strong></p>

<pre><strong>Input:</strong> s = "bbbbb"
<strong>Output:</strong> 1
<strong>Explanation:</strong> The answer is "b", with the length of 1.
</pre>

<p><strong class="example">Example 3:</strong></p>

<pre><strong>Input:</strong> s = "pwwkew"
<strong>Output:</strong> 3
<strong>Explanation:</strong> The answer is "wke", with the length of 3.
Notice that the answer must be a substring, "pwke" is a subsequence and not a substring.
</pre>

<p>&nbsp;</p>
<p><strong>Constraints:</strong></p>

<ul>
	<li><code>0 &lt;= s.length &lt;= 5 * 10<sup>4</sup></code></li>
	<li><code>s</code> consists of English letters, digits, symbols and spaces.</li>
</ul>
</div>

```
def lengthOfLongestSubstring(s):
    i = 0
    longest_substring_final = ''
    longest_substring = ''
    longest_substring_obj = {}
    
    while i < len(s):
        if s[i] not in longest_substring_obj:
            longest_substring_obj[s[i]] = ''
            longest_substring += s[i]   
        else:
            if len(longest_substring) > len(longest_substring_final):
                longest_substring_final = longest_substring
            while True:
                first_char = longest_substring[0]
                longest_substring = longest_substring[1:]
                del longest_substring_obj[first_char]  
                if first_char == s[i]:
                    break 
            longest_substring += s[i] 
            longest_substring_obj[s[i]] = ''
        
        i += 1
        
    if len(longest_substring) > len(longest_substring_final):
        longest_substring_final = longest_substring
    
    return len(longest_substring_final)

lengthOfLongestSubstring('abcabccbb')
```
