**[From Leetcode: 914. X of a Kind in a Deck of Cards](https://leetcode.com/problems/x-of-a-kind-in-a-deck-of-cards)**

<div class="xFUwe" data-track-load="description_content"><p>You are given an integer array <code>deck</code> where <code>deck[i]</code> represents the number written on the <code>i<sup>th</sup></code> card.</p>

<p>Partition the cards into <strong>one or more groups</strong> such that:</p>

<ul>
	<li>Each group has <strong>exactly</strong> <code>x</code> cards where <code>x &gt; 1</code>, and</li>
	<li>All the cards in one group have the same integer written on them.</li>
</ul>

<p>Return <code>true</code><em> if such partition is possible, or </em><code>false</code><em> otherwise</em>.</p>

<p>&nbsp;</p>
<p><strong class="example">Example 1:</strong></p>

<pre><strong>Input:</strong> deck = [1,2,3,4,4,3,2,1]
<strong>Output:</strong> true
<strong>Explanation</strong>: Possible partition [1,1],[2,2],[3,3],[4,4].
</pre>

<p><strong class="example">Example 2:</strong></p>

<pre><strong>Input:</strong> deck = [1,1,1,2,2,2,3,3]
<strong>Output:</strong> false
<strong>Explanation</strong>: No possible partition.
</pre>

<p>&nbsp;</p>
<p><strong>Constraints:</strong></p>

<ul>
	<li><code>1 &lt;= deck.length &lt;= 10<sup>4</sup></code></li>
	<li><code>0 &lt;= deck[i] &lt; 10<sup>4</sup></code></li>
</ul>
</div>

```
class Solution(object):
    def hasGroupsSizeX(self, deck):
        """
        :type deck: List[int]
        :rtype: bool
        """
        obj_ = {}
        for num in deck:
            if num not in obj_:
                obj_[num] = 1
            else:
                obj_[num] += 1

        min_num_ = 2**31
        odd_flag = False
        for key in obj_.keys():
            if obj_[key] < min_num_:
                min_num_ = obj_[key]

            if obj_[key] % 2 != 0:
                odd_flag = True

        if min_num_ == 1:
            return False

        factor = 2
        if odd_flag:
            for i in range(3, min_num_ + 1, 2):
                for key in obj_.keys():
                    if obj_[key] % i == 0:
                        factor = i
                    else:
                        break
                else:
                    break

        for key in obj_.keys():
            if obj_[key] % factor != 0:
                return False

        return True
```
