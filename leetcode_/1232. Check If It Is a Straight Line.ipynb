**[From Leetcode: 1232. Check If It Is a Straight Line](https://leetcode.com/problems/check-if-it-is-a-straight-line)**

<div class="xFUwe" data-track-load="description_content"><p>You are given an array&nbsp;<code>coordinates</code>, <code>coordinates[i] = [x, y]</code>, where <code>[x, y]</code> represents the coordinate of a point. Check if these points&nbsp;make a straight line in the XY plane.</p>

<p>&nbsp;</p>

<p>&nbsp;</p>
<p><strong>Constraints:</strong></p>

<ul>
	<li><code>2 &lt;=&nbsp;coordinates.length &lt;= 1000</code></li>
	<li><code>coordinates[i].length == 2</code></li>
	<li><code>-10^4 &lt;=&nbsp;coordinates[i][0],&nbsp;coordinates[i][1] &lt;= 10^4</code></li>
	<li><code>coordinates</code>&nbsp;contains no duplicate point.</li>
</ul>
</div>

```
class Solution(object):
    def checkStraightLine(self, coordinates):
        """
        :type coordinates: List[List[int]]
        :rtype: bool
        """
        if len(coordinates) <= 2: return True
        a = coordinates[0][0] - coordinates[1][0]
        b = coordinates[0][1] - coordinates[1][1]
        for i in range(2, len(coordinates)):
            a_ = coordinates[0][0] - coordinates[i][0]
            b_ = coordinates[0][1] - coordinates[i][1]
            if b == 0 and b_ == 0: continue
            if b * b_ == 0: return False
            if 1.0 * a / b != 1.0 * a_ / b_: return False
            # print(a_, b_)

        return True
```
