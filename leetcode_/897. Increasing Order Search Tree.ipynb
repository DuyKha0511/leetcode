**[From Leetcode: 897. Increasing Order Search Tree](https://leetcode.com/problems/increasing-order-search-tree)**

<div class="xFUwe" data-track-load="description_content"><p>Given the <code>root</code> of a binary search tree, rearrange the tree in <strong>in-order</strong> so that the leftmost node in the tree is now the root of the tree, and every node has no left child and only one right child.</p>

<ul>
	<li>The number of nodes in the given tree will be in the range <code>[1, 100]</code>.</li>
	<li><code>0 &lt;= Node.val &lt;= 1000</code></li>

```
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution(object):
    def increasingBST(self, root):
        """
        :type root: TreeNode
        :rtype: TreeNode
        """
        nodes = self.inorder(root, [])
        node = TreeNode(val=nodes[0])
        temp = node
        for i in range(1, len(nodes)):
            temp.right = TreeNode(val=nodes[i])
            temp = temp.right

        return node

    def inorder(self, root, nodes):
        if root is not None:
            self.inorder(root.left, nodes)
            nodes.append(root.val)
            self.inorder(root.right, nodes)

        return nodes
```
