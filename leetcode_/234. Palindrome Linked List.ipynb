<ul>

<p>&nbsp;</p>
<strong>Follow up:</strong> Could you do it in <code>O(n)</code> time and <code>O(1)</code> space?</div>

```
class ListNode(object):
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def isPalindrome(head):
    """
    :type head: ListNode
    :rtype: bool
    """
    if head == None: return None
    temp = head
    reverse_list = None
    while temp:
        new_node = ListNode(temp.val, reverse_list)
        reverse_list = new_node
        temp = temp.next

    temp1 = reverse_list
    temp2 = head
    while temp1:
        if temp1.val != temp2.val:
            return False
        temp1 = temp1.next
        temp2 = temp2.next

    return True
```
